```python
from bs4 import BeautifulSoup
import requests
import pandas as pd



covid_url="https://www.worldometers.info/coronavirus/"

page_covid = requests.get(covid_url).text

soup = BeautifulSoup(page_covid, "lxml")
```

```python
table = soup.find("table", attrs={"id": "main_table_countries_today"})

head_covid = table.thead.find_all("tr")
columns = []
for th in head_covid[0].find_all("th"):
    columns.append(th.text.replace("\n","").strip())

body_covid = table.tbody.find_all("tr") 
val = []
for r in range(1,len(body_covid)):
    rows = [] 
    for tr in body_covid[r].find_all("td"):
        rows.append(tr.text.replace("\n","").strip())
    val.append(rows)
covid_data = pd.DataFrame(val,columns=columns)
covid_data = covid_data[covid_data["#"]!=""].reset_index(drop=True)
covid_data.drop(columns=['#', 'Tot\xa0Cases/1M pop', 'Deaths/1M pop', 'Tests/1M pop','1 Caseevery X ppl', '1 Deathevery X ppl', '1 Testevery X ppl', 'New Cases/1M pop', 'New Deaths/1M pop', 'Active Cases/1M pop'], inplace=True)
covid_data = covid_data.drop_duplicates(subset = ["Country,Other","Continent"])
print("Shape: ",covid_data.shape)
covid_data.head(15)
```

```python
covid_data.to_csv("dataset/covid_data.csv", sep='\t')
```
